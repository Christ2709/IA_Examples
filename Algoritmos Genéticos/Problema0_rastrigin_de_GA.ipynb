##Rastrigin para 5 variáveis. Limite: [-5, 5]
Ótimo conhecido em x = [0, ..., 0] e avaliação = 0

```
from scipy.optimize import differential_evolution
import math
```

```
def rastrigin(solution):
  result = 0
  for var in range(len(solution)):
    result += solution[var]**2 - 10*math.cos(2*math.pi*solution[var]) +10
  return result
```

limites = [ (lb, up), ..., (lb, ub) ]

```
nvars = 5 #variables de decisión
bounds = nvars*[(-5,5)] #rango de solución
bounds
```

```
result = differential_evolution(rastrigin, bounds)
```

```
result
```

```
print("Melhor solução: ", result.x)
```

```
print("Melhor avaliação: ", result.fun)
```
