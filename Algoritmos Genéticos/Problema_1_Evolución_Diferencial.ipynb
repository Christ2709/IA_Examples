# Evolução Diferencial em Python (Scipy)

## Problema 1

Encontre dois números positivos onde a soma do primero com o dobro do segundo resulte em 100 e o produto entre os dois seja máximo. 

$x+2y=100$

$x*y= Max$

##Rastrigin para 5 variáveis. Limite: [0, 100]


```
from scipy.optimize import differential_evolution, NonlinearConstraint, Bounds
import numpy as np
import math
```

```
def constraint(solution):
    return np.array(solution[0] + 2*solution[1])
```

```
# the sum of x[0] and x[1] must be 100
nlc = NonlinearConstraint(constraint, lb=100, ub=100)
```

```
bounds = 2*[(0,100)]
bounds
```

```
def evaluation(solution):
  return -(solution[0]*solution[1]) # inverter sinal: max
```

```
result = differential_evolution(evaluation, bounds, constraints=(nlc))
```

```
print("Melhor solução: ", result.x)
```

```
print("Melhor avaliação: ", -result.fun)
```
